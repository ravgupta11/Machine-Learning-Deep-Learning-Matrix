```
import numpy as np
import pandas as pd

def normalize(df):
  for col in df.columns[1:-1]:
    mx = df[col].max()
    mn = df[col].min()
    df[col] = (df[col] - mn) / (mx - mn)  ## 1 2 3 4 5 6 7 8 9 9 10 (subtract minimum) -> 0 1 2 3 4 5 6 7 8 8 9 > (divide by max size = mx - mn) 0 1/9 2/9 ... 9/9 = 0 1/9 2/9 3/9 ... 1
  return df

df = pd.read_csv('Admission_Predict_Train.csv', delimiter = ',')
df = normalize(df)
df
```

```

alpha = 0.1
epochs = 1000000
theta = np.ones([1, 8])
def sigmoid(x):
    return np.where(x >= 0, 
                    1 / (1 + np.exp(-x)), 
                    np.exp(x) / (1 + np.exp(x)))

def compute_probability(X, theta):
  p_w_b = sigmoid((X @ theta.T))
  q_w_b = 1.00 - p_w_b
  return (p_w_b, q_w_b)

def compute_gradient(X, p,y):
  return np.sum((y - p) * X, axis = 0) / float(len(y))

def compute_likelihood(y, p, q):
  return np.sum(y * np.log(p) + (1.00-y)* np.log(q)) / float(len(y))

def gradient_descent(X, theta, y):
  for i in range(epochs):
    p, q = compute_probability(X, theta)
    gradient = compute_gradient(X, p,y)
    theta = theta + alpha * gradient;
    p, q = compute_probability(X, theta)
    cost = compute_likelihood(y, p, q)
  return (theta, cost)

theta, cost = gradient_descent(X, theta, y)
```
